```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
bronx_url = 'https://newyork.craigslist.org/d/apartments-housing-for-rent/search/brx/apa?'

brooklyn_url = 'https://newyork.craigslist.org/d/apartments-housing-for-rent/search/brk/apa?'

queens_url = 'https://newyork.craigslist.org/d/apartments-housing-for-rent/search/que/apa?'

manhattan_url = 'https://newyork.craigslist.org/d/apartments-housing-for-rent/search/mnh/apa?'

staten_url = 'https://newyork.craigslist.org/d/apartments-housing-for-rent/search/stn/apa?'
```

```python
import os
from selenium import webdriver
import time
#provide path to the chromedriver
chromedriver = "/Applications/chromedriver"
os.environ["webdriver.chrome.driver"] = \
    chromedriver
driver = webdriver.Chrome(chromedriver)

#using selenium to get the html of the 1st bronx results page
driver.get(bronx_url)
bronx_html = driver.page_source
time.sleep(2)
bronx_front_soup = bs(bronx_html)

import math
def url_list_maker(url_base, front_soup):
    url_list = []
    total_count = int(front_soup.find('span', {'class':'totalcount'}).text)
    total_pages = math.ceil(total_count / 120)
    for i in range(total_pages):
        counter = i * 120
        url_list.append(url_base+str('s=')+str(counter))
    return url_list

bronx_urls = url_list_maker(bronx_url, bronx_front_soup)

#counting the pages:
print(len(bronx_urls))
```

```python
#using selenium to get the html of the 1st brooklyn results page
driver.get(brooklyn_url)
brooklyn_html = driver.page_source
time.sleep(2)
brooklyn_front_soup = bs(brooklyn_html)

#making a list of pages to scrape
brooklyn_urls = url_list_maker(brooklyn_url, brooklyn_front_soup)

#counting the pages:
print(len(brooklyn_urls))
```

```python
#using selenium to get the html of the 1st queens results page
driver.get(queens_url)
queens_html = driver.page_source
time.sleep(2)
queens_front_soup = bs(queens_html)

#making a list of pages to scrape
queens_urls = url_list_maker(queens_url, queens_front_soup)

#counting the pages:
print(len(queens_urls))
```

```python
#using selenium to get the html of the 1st Manhattan results page
driver.get(manhattan_url)
manhattan_html = driver.page_source
time.sleep(2)
manhattan_front_soup = bs(manhattan_html)

#making a list of pages to scrape
manhattan_urls = url_list_maker(manhattan_url, manhattan_front_soup)

#counting the pages:
print(len(manhattan_urls))
```

```python
#using selenium to get the html of the 1st Staten Island results page
driver.get(staten_url)
staten_html = driver.page_source
time.sleep(2)
staten_front_soup = bs(staten_html)

#making a list of pages to scrape
staten_urls = url_list_maker(staten_url, staten_front_soup)

#counting the pages:
print(len(staten_urls))
```

```python
#function to scrape each page of search results for the individual post 
def page_scraper(url_list):
    borough_list = []
    for url in url_list:
        driver.get(url)
        page = driver.page_source
        time.sleep(3)
        soup = bs(page)
        for item in soup.find_all('h3', class_='result-heading'):
            borough_list.append(item.find('a')['href'])
    return borough_list
```

```python
brx_posts = page_scraper(bronx_urls)

print(len(brx_posts))
```

```python
brk_posts = page_scraper(brooklyn_urls)

print(len(brk_posts))
```

```python
mnh_posts = page_scraper(manhattan_urls)

print(len(mnh_posts))
```

```python
que_posts = page_scraper(queens_urls)

print(len(que_posts))
```

```python
stn_posts = page_scraper(staten_urls)

print(len(stn_posts))
```

```python
import random

def post_scraper(borough_list):
    listing_info = []
    for url in borough_list:
        driver.get(url)
        page = driver.page_source
        time.sleep(random.randint(1, 3))
        post_info = bs(page)
        #price
        try:
            price = post_info.find('span', class_='price').text
        except:
            price = 'n/a'
        #listing title
        try:
            title = post_info.find('span', id="titletextonly").text
        except:
            title = 'n/a'
        #location
        try: 
            location = post_info.find('span', {'class': 'postingtitletext'}).find('small').text
        except:
            location = 'n/a'
        #bedrooms
        try:
            br = post_info.find('span', class_='shared-line-bubble').find_all('b')[0].text
        except:
            br = 'n/a'
        #bathrooms
        try:
            ba = post_info.find('span', class_='shared-line-bubble').find_all('b')[1].text
        except:
            ba = 'n/a'
        #sq feet
        try:
            sqft = post_info.find_all('span', class_='shared-line-bubble')[1].find_all('b')[0].text
        except:
            sqft = 'n/a'
        #application fee
        try:
            app_fee = post_info.find(text='application fee details: ').next.text
        except:
            app_fee = 'n/a'
        #brokers fee
        try:
            brok_fee = post_info.find(text='broker fee details: ').next.text
        except:
            brok_fee = 'n/a'
        #cats allowed
        if 'cats are OK - purrr'in post_info.text:
            cats = True
        else:
            cats = 'n/a'
        #dogs allowed
        if 'dogs are OK - wooof'in post_info.text:
            dogs = True
        else:
            dogs = 'n/a'
        #laundry
        if 'no laundry on site' in post_info.text:
            wd = 'none'
        elif 'laundry on site' in post_info.text:
            wd = 'on site'
        elif 'laundry in bldg' in post_info.text:
            wd = 'in bldg'
        elif 'w/d in unit' in post_info.text:
            wd = 'in unit'
        elif 'w/d hookups' in post_info.text:
            wd = 'hookups'
        else:
            wd = 'n/a'
        #parking
        if 'no parking' in post_info.text:
            parking = 'none'
        elif 'off-street parking' in post_info.text:
            parking = 'off-street'
        elif 'street parking' in post_info.text:
            parking = 'street'
        elif ('attached garage' in post_info.text) or ('detached garage' in post_info.text):
            parking = 'garage'
        elif 'valet parking' in post_info.text:
            parking = 'valet'
        elif 'carport' in post_info.text:
            parking = 'carport'
        else:
            parking = 'n/a'

        listing_info.append({
                'PRICE': price,
                'TITLE': title,
                'LOCATION': location, 
                'BR': br, 
                'BA': ba, 
                'SQFT': sqft, 
                'APP_FEE': app_fee, 
                'BROK_FEE': brok_fee,
                'CATS': cats,
                'DOGS': dogs,
                'WD': wd,
                'PARKING': parking
                    })

    return listing_info
```
